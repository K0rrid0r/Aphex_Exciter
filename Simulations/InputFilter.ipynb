```python
import numpy as np
import scipy.signal as signal
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
_C_1 = 47e-6
_C_2 = 56e-12
_R_1 = 15000
_R_2 = 22000
_R_3 = 10000
```

```python
b1 = _R_1 * (_R_2 + _R_3) * _C_1
b0 = _R_2 + _R_3

a2 = _C_1 * _C_2 * (_R_2 + _R_3 + 1)
a1 = b1
a0 = b0

w, H = signal.freqs ([0, b1, b0], [a2, a1, a0])
plt.figure()
plt.semilogx (w / (2 * np.pi), 20*np.log10 (np.abs (H)))
```
