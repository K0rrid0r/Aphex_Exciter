```python
import numpy as np
import scipy.signal as signal
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
R_1 = 20000
R_2 = 10000
C = 20e-12
```

## Input Filter

```python
a1_i = R_1 * C
a0_i = 1

b1_i = 0
b0_i = 1

w, H = signal.freqs ([b1_i, b0_i], [a1_i, a0_i])
plt.figure()
plt.semilogx (w / (2 * np.pi), 20 * np.log10 (np.abs (H)))
plt.xlim ((20, 80000))
```

## Control Filter

```python
a1_c = R_1 * R_2 * C
a0_c = R_2

b1_c = 0
b0_c = R_1

w, H_c = signal.freqs ([b1_c, b0_c], [a1_c, a0_c])
plt.figure()
plt.semilogx (w / (2 * np.pi), 20 * np.log10 (np.abs (H_c)))
plt.xlim ((20, 80000))
plt.ylim ((4, 8))
plt.axhline (6, color='r')
```
